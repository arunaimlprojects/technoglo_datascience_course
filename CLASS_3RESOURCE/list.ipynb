Converting a List to String

```python
my_list = [1, "apple", 3.14, True]

# Converting List to String
list_to_string = str(my_list)

print(list_to_string)
# Output: "[1, 'apple', 3.14, True]"
```

```python
type(list_to_string)
```

Convert String to List Using split()

```python
my_string = "apple banana cherry"
# Converting String to List by splitting on space
my_list = my_string.split()

print(my_list)
# Output: ['apple', 'banana', 'cherry']
```

```python
#2. List Deletion Functions:
# Example:
my_list = [10, 20, 30, 40, 50]
del my_list[2]  # Index 2 se element delete karega (30 ko)
print(my_list)  # Output: [10, 20, 40, 50]

del my_list  # Whole list ko delete karega
# print(my_list)  # Error, as my_list is deleted
```

```python
#Using remove()
# Example:
my_list = [10, 20, 30, 40, 50]
my_list.remove(30)  # 30 ko remove karega
print(my_list)  # Output: [10, 20, 40, 50]

# Agar element list mein nahi ho to error aayegi
# my_list.remove(60)  # ValueError: list.remove(x): x not in list
```

```python
#Using pop()
# Example:
my_list = [10, 20, 30, 40, 50]
last_item = my_list.pop()  # Last item ko remove karega
print(last_item)  # Output: 50
print(my_list)    # Output: [10, 20, 30, 40]

second_item = my_list.pop(1)  # Index 1 (20) ko remove karega
print(second_item)  # Output: 20
print(my_list)      # Output: [10, 30, 40]
```

```python
#Using clear()
# Example:
my_list = [10, 20, 30, 40, 50]
my_list.clear()  # All elements will be removed
print(my_list)   # Output: []
```

```python
#append()
# Example:
my_list = [10, 20, 30]
my_list.append(40)
print(my_list)  # Output: [10, 20, 30, 40]
```

```python
#insert()
# Example:
my_list = [10, 20, 30]
my_list.insert(1, 15)  # Index 1 pe 15 insert karega
print(my_list)  # Output: [10, 15, 20, 30]
```

```python
#del
# Example:
my_list = [10, 20, 30, 40]
del my_list[1]  # Index 1 se element (20) ko remove karega
print(my_list)   # Output: [10, 30, 40]

# Entire list ko delete karne ke liye
del my_list  # List ko delete kar dega
# print(my_list)  # Error, because the list no longer exists
```

```python
#sorted 
# Example: Sorting list in ascending order
my_list = [5, 2, 9, 1, 5, 6]
sorted_list = sorted(my_list)
print(sorted_list)  # Output: [1, 2, 5, 5, 6, 9]
```

```python
#extend()
my_list = [1, 2, 3]
my_list.extend([4, 5, 6])  # Adding multiple elements
print(my_list)  # Output: [1, 2, 3, 4, 5, 6]
```

```python
#1. Concatenation (+ Operator)
# Example: Concatenation using + operator
list1 = [1, 2, 3]
list2 = [4, 5, 6]
result = list1 + list2
print(result)  # Output: [1, 2, 3, 4, 5, 6]
```

```python
#Repetition (* Operator)
# Example: Repeating a list using * operator
my_list = [1, 2, 3]
result = my_list * 3
print(result)  # Output: [1, 2, 3, 1, 2, 3, 1, 2, 3]
```

```python
# Membership Test (in and not in Operators)
# Example: Checking if an element exists in a list
my_list = [1, 2, 3, 4, 5]
print(3 in my_list)  # Output: True (3 is in the list)
print(6 not in my_list)  # Output: True (6 is not in the list)
```

```python
#List Length (len() function)
# Example: Getting the length of a list
my_list = [10, 20, 30, 40, 50]
print(len(my_list))  # Output: 5
```

```python
#Reversing (reverse() Method)
# Example: Reversing a list
my_list = [10, 20, 30, 40]
my_list.reverse()
print(my_list)  # Output: [40, 30, 20, 10]
```

```python
#Min, Max and Sum (min(), max(), sum())
# Example: Finding min, max, and sum of elements in a list
my_list = [10, 20, 30, 40, 50]
print(min(my_list))  # Output: 10 (minimum value)
print(max(my_list))  # Output: 50 (maximum value)
print(sum(my_list))  # Output: 150 (sum of elements)
```

```python
# Example: Using a for loop to remove duplicates
my_list = [1, 2, 2, 3, 4, 4, 5]
unique_list = []
for item in my_list:
    if item not in unique_list:
        unique_list.append(item)
print(unique_list)  # Output: [1, 2, 3, 4, 5]
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def perform_eda(dataset_name):
    # Load dataset
    data = sns.load_dataset(dataset_name)
    
    # First 5 rows
    print("First 5 rows of the dataset:")
    print(data.head())

    # Basic statistical summary
    print("\nStatistical summary of the dataset:")
    print(data.describe())
    
    # Checking for missing values
    print("\nMissing values in the dataset:")
    print(data.isnull().sum())
    
    # Correlation heatmap - only numerical columns
    numerical_data = data.select_dtypes(include=[np.number])
    plt.figure(figsize=(8,6))
    sns.heatmap(numerical_data.corr(), annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
    plt.title(f"Correlation Heatmap for {dataset_name}")
    plt.show()

    # Distribution plot for numerical features
    numerical_cols = numerical_data.columns
    for col in numerical_cols:
        plt.figure(figsize=(6,4))
        sns.histplot(data[col], kde=True, bins=20)
        plt.title(f'Distribution of {col}')
        plt.show()
    
    # Box plot for numerical features
    for col in numerical_cols:
        plt.figure(figsize=(6,4))
        sns.boxplot(x=data[col])
        plt.title(f'Box Plot of {col}')
        plt.show()

    # Pairplot for understanding relationships
    sns.pairplot(numerical_data)
    plt.suptitle(f"Pairplot for {dataset_name}", y=1.02)
    plt.show()

# Load the iris dataset and perform EDA
perform_eda('iris')
```

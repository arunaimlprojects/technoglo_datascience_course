
# String in Python Quiz

### 1. Which of the following is an example of a string in Python?
**Answer:** 1. `'Hello, World!'`

---

### 2. How do you access the first character of a string in Python?
**Answer:** 1. `s[0]`

---

### 3. What will be the output of the following code snippet?
```python
s = 'Python'
print(s[1:4])
```
**Answer:** 1. `yth`

---

### 4. Which method is used to find the index of the first occurrence of a substring within a string in Python?
**Answer:** 2. `find()`

---

### 5. What will be the output of the following code snippet?
```python
s = 'hello'
print(s.upper())
```
**Answer:** 2. `HELLO`

---

### 6. Which of the following is NOT a valid way to concatenate two strings in Python?
**Answer:** 1. `s1 . s2`

---

### 7. What will be the output of the following code snippet?
```python
s = 'python'
print(s.replace('p', 'P', 1))
```
**Answer:** 4. `Ppython`

---

### 8. Which of the following is the correct way to check if a string contains only alphabets in Python?
**Answer:** 1. `s.isalpha()`

---

### 9. What will be the output of the following code snippet?
```python
s = 'apple,orange,mango'
print(s.split(','))
```
**Answer:** 2. `['apple', 'orange', 'mango']`

---

### 10. Which method is used to remove leading and trailing whitespaces from a string in Python?
**Answer:** 2. `strip()`

---

### 11. What will be the value of the expression `'3' + '4'` in Python?
**Answer:** 1. `34`

---

### 12. Which of the following is NOT a valid method to format a string in Python?
**Answer:** 4. `str.format()`

---

### 13. What will be the output of the following code snippet?
```python
s = 'hello'
print(s.count('l'))
```
**Answer:** 2. `2`
